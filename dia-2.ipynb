```python
#Challenge 2


def temperatura(temperatura_ambiental):
    print(f'La temperatura ambiental es: {temperatura_ambiental}')

    if temperatura_ambiental >= -10 and temperatura_ambiental <= 13:
        print("Legalmente Ro'y koko")
    elif temperatura_ambiental >= 14 and temperatura_ambiental <= 23 :
        print("Refreska'i")
    elif temperatura_ambiental >= 24 and temperatura_ambiental <= 28:
        print("Oikoite")
    elif temperatura_ambiental >= 29 and temperatura_ambiental <= 45 :
        print("Hakurooooo")
    else: 
        print("Ñamano mba'e kpeeeee")

temperatura(int(input('Por favor ingrese la temperatura en grados Celsius')))



```

```python
#Challenge 7
frase = input('Ingrese su frase')
contador = 0
for caracteres in frase:
    if caracteres == 'o' or caracteres == 'O':
        contador += 1
print(f'Su frase "{frase}" detecta la letra o. Cantidad de o en esta frase: {contador}')
        
```
