```python
numero_usuario = 10

while True: 
    numero_ingresado = int(input('ingrese el numero'))
    if numero_ingresado == numero_usuario:
        print(f'El numero {numero_ingresado} ingresado es el correcto')
        break 

    else:
        print('El numero ingresado es incorrecto.')
```

```python
numero_a_encontrar = 100
intentos = 5
while intentos > 0:
    numero_a_ingresar = int(input('Adivine el numero'))
    
    if numero_a_ingresar == numero_a_encontrar:
        print(f'El numero {numero_a_ingresar} es el correcto!!')
    break
elif numero_a_ingresar != numero_a_encontrar
    print(f'El numero {numero_a_ingresar} es incorrecto, vuelve a intentar.')
else:
     intentos -= 1
     print('Intentos restantes: ', intentos)
     if intentos == 0:
          print('Ya no te quedan mas intentos. Prueba mas tarde')

```

```python
numero_secreto = 7
intentos = 3

while intentos > 0:
    numero = int(input("Adivina el número: "))

    if numero == numero_secreto:
        print("¡Correcto! Adivinaste el número ")
        break
    else:
        intentos = intentos - 1

        if numero < numero_secreto:
            print("El número a adivinar es mayor")
        else:
            print("El número a adivinar es menor")

        print("Te quedan", intentos, "intentos")

if intentos == 0:
    print("Se acabaron los intentos ")
```
