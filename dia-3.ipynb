```python
#challenge 3

palabra = len(input('Ingrese su palabra'))
numero = int(input('Ingrese su numero'))

def comparar_palabras(palabra, numero):
    if palabra == numero:
        return True
    else:
        return False
resultado = comparar_palabras(palabra, numero)
print(f'El resultado es {resultado}')

```

```python
palabra_usuario = input('Ingrese una palabra')
numero_usuario = int(input('Ingrese un numero'))

def cantidad_letras(word, number):

    numero_de_letras = len(word)

    if numero_de_letras == number:
        return True

    else:
        return False

resultado = cantidad_letras(palabra_usuario, numero_usuario)
print (f'El resultado es {resultado}')
```
