```python
#Ejemplo operaciones
cantidad_de_naranjas = 10
cantidad_de_manzanas = 7
cantidad_de_frutas = cantidad_de_naranjas + cantidad_de_manzanas
print('El total de las frutas es de: ', cantidad_de_frutas)

frutas_vendidas = 12
frutas_restantes = cantidad_de_frutas - frutas_vendidas
print('La cantidad de frutas restantes es: ', frutas_restantes)
```

```python
#Input y operaciones de division y modulo
cant_caramelos = int(input('Ingrese la cantidad de caramelos:'))
cant_niños = int(input('Ingrese la cantidad de niños: '))
caramelos_repartidos = cant_caramelos / cant_niños
print(f'El total de caramelos repartidos es: {caramelos_repartidos}')

caramelos_restantes= cant_caramelos % cant_niños
print(f'La cantidad de caramelos restantes es: {caramelos_restantes}')
```
